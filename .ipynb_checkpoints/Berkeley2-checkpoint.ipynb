```python
import numpy as np
import pandas as pd
import csv
```

```python
arr = []
with open("data.csv", newline='' ) as csvfile:
    for row in csvfile.readlines()[1:]:
        row = row.rstrip()
        arr.append(row.split(','))
       #print(row)

iui = []
for ar in arr:
    iui.append(ar[0] + " " +  ar[1])
#print(iui)
print(len(set(iui)))
```

```python
dataFrame = pd.DataFrame(arr, columns = ["gene_id","spot_id","time","spot_fluorescence"])
dataFrame
```
